# Functions to obtain player's different markets probabilities

## First function to call and save a DF from csv file

```python
def load_file(filename):
    """This function loads the inital file and rename different columns and returns df
    
    """
    df = pd.read_csv(filename)
    df.rename(columns={'Unnamed: 6': 'Location','Unnamed: 8':'Result'},inplace=True)
    # replace NAN values in Location column by 'vs' meaning home game
    df['Location'].fillna('vs',inplace=True)
    
    return df
    
    
```

### run the following cell

```python
df1 = load_file('basic_per_game_player_stats_2013_2018.csv')
```

## Second function to call, assign the two outputs


```python
def teams_matchup(df,tm1,tm2):
    """
    This function filters the original DF to obtain a single DF showing the matchups of two given teams.
    Returns a filtered df and a list of players involved
    inputs: df-original DF
            tm1 and tm2 - both teams in 'AAA' form
    output: game_df - df filtered by teams
            players - list of players involved in the match up
    """
    ## creating a list of teams in DF
    teams =list(df.Tm.unique())
    if (tm1 not in teams) | (tm2 not in teams):
        print("Team not found in data")
    else:

        # Slicing df with the conditionals for teams match-up  
        game_df = df[((df['Tm'] == tm1) & (df['Opp'] == tm2)) |((df['Tm'] == tm2) & (df['Opp'] == tm1))]
        # creting players list
        game_players = list(game_df['Player'].unique())
        print(f'There are {len(game_players)} different players involved in {tm1} vs {tm2} match.')
        #print(game_players)
    
    return game_df,game_players
    
    
```

## Third function to be called

```python
def players_stats(player,market,df,game_df,plots):
    """
    This function filters the df by a unique player, obtained from the players list in a given match up, and
    provides insights of player's stats. The function can also show correlation plots between minutes played
    marktet analysed.
    inputs: player - name and code of a player, taken from palyers list prev. created
            market -  betting market to be analysed:
                        'PTS' total points
                        '3P' total 3 pointers
                        'AST' total assists 
                        'TRB' total rebounds
                        'STL' total steals
                        'BLK' total blocks
            df - original df
            game_df - df filtered by teams
            plots - If TRUE, the function will display the plots, FALSE to avoid display
    outputs: This function displays correlation plots for given player between min played and market, 
            also prints statements of players stats.
            
            Thw two df used can be returned in oprder to check the stats on each player
            player_df - df of player's total appereances
            player_opp - df of player vs Opp team
    """
    # filtering original df for player appereances
    player_df = df[df['Player'] == player]
    # calculating mean  min played in all apps and mean market total
    mean_min = round(player_df.MP.mean(),2)
    mean_market = round(player_df[market].mean(),2)
    
    print(f'The average minutes played for {player} is: \n {mean_min}')
    print(f'The average {market} per game for {player} is: \n {mean_market}')
    
    # filtering match up df for player vs given Opp team
    player_opp = game_df[game_df['Player'] == player]
    # calculating mean  min played and mean market vs given Opp team
    mean_min_opp = round(player_opp.MP.mean(),2)
    mean_market_opp = round(player_opp[market].mean(),2)
    opp_team = player_opp.Opp.unique()[0]
    print(f'The average minutes played for {player} against {opp_team} is: \n {mean_min_opp}')
    print(f'The average {market} per game for {player} against {opp_team} is: \n {mean_market_opp}')
    print('----------------------------------------------------------------------- \n')
    if plots == True:
        
        #plots correlation player min played vs market totals
        #axes = plt.subplots(1,2,figsize=(8,8))
        sns.lmplot(x='MP',y=market,data=player_df,hue='GS',)#ax=axes[0,0])
        plt.title(f'{player} correlation between minutes played and {market} in all apps.')
    
        sns.lmplot(x='MP',y=market,data=player_opp,hue='GS',)#ax=axes[0,1])
        plt.title(f'{player} correlation between minutes played and {market} vs {opp_team}.')

        plt.show()
        
    # Calculate combined stats player all apps and player vs Opp team
    mean_min_comb = (mean_min + mean_min_opp) / 2
    mean_market_comb = (mean_market + mean_market_opp) / 2
    
    # Total number of games for given player 
    number_games = player_df.shape[0]
    print(f'The total number of games played by {player} is: \n {number_games}')
    # Total number of times given player went over in all appereances
    total_over_market = player_df[player_df[market] > mean_market].shape[0]
    print(f'Number of times {player} was over the average {market} is: \n {total_over_market} ')
    # Number of times the player was a starter in mean_market
    starter = player_df[player_df[market] > mean_market]['GS'].sum()
    print(f'The number of times player was a starter: \n {starter}')
    # Prob of player going over  considering all appereances
    total_over_proba = total_over_market / number_games
    print(f'The probability of {player} going over in {market} is: \n {round(total_over_proba,2)}')
    print('----------------------------------------------------------------------- \n')
    print(f'The suggested {market} Over/Under value for {player} vs {opp_team} is: \n {round(mean_market_comb,2)}')
    print('----------------------------------------------------------------------- \n')   
    # Total number of games for given player vs Opp team
    number_games_opp = player_opp.shape[0]
    print(f'The number of games {player} has against {opp_team} is: \n {number_games_opp}')
    # Total number of times given player went over vs Opp
    games_over_market = player_opp[player_opp[market] > mean_market_comb].shape[0]
    print(f'Number of times {player} was over the average {market} vs {opp_team} is: \n {games_over_market}')
    # Number of times the player was a starter in games_over_market
    starter_opp = player_opp[player_opp[market] > mean_market]['GS'].sum()
    print(f'The number of times player was a starter: \n {starter_opp}')
    # Probability of player getting over vs Opp
    over_proba = games_over_market / number_games_opp
    print(f'The probability of {player} going over in {market} vs {opp_team} is: \n {round(over_proba,2)}')
    print('----------------------------------------------------------------------- \n')   
    # Number of times a player is a starter in all appereances
    starter_app = player_df[player_df['GS'] == 1 ].shape[0]
    print(f'Number of times {player} was a starter in all appereances is: \n {starter_app}')
    # Probability of player being a starter
    starter_proba = starter_app / number_games
    print(f'Probability of {player} being a starter in all apperenaces is: \n {round(starter_proba,2)}')
    
    #for testing return both dataframes
    # run the next line instead
    return player_df,player_opp
#     return
    
    
```

## Future work
There are some elements I would spent more time on and probably make this process more efficient. During the process I filtered the data according to 2 teams matches against each other and generating a list of players that have been involed in this matches, however I am aware that all the players listed are not necessarily still on the teams so it is something to be considered.

There seems to be a correlation between the minutes played and the different markets, simply if a player has more game time he has the chance to score more points. On the same note, if a player is a starter he probably has the opportunity to increase his game time and therefore affect the probability of going over. So, being a started potentially will aid the players stats and it could be included in the analysis.

Also, as it happens in many sports, at the bneginning of the season a player is likely to under perform (coming from preseason or just not use to new team or coach). For this reason adding the month of when a match is happening can also have an effect on the player stats, if it is a play-off mnonthm (end of season) a player could be performing better than at the begining of the season.

In the case of rookies in which there is no available data, one of the solution could be to gather the data of players rookie seasons and considere an average of those stats. Player's stats from NCAA could also be used, however, I think the level and the ammount of games in NCAA is not a very good comparisson with NBA.
